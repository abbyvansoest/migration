```python
import pandas as pd
```

```python
base_graph = pd.read_csv('../results/link_centrality_measures.csv')
base_graph = base_graph[base_graph['Edge Betweenness'] > 0]
base_graph = base_graph.sort_values(by='Edge Betweenness', ascending=False).reset_index(drop=True)
base_graph['Original Rank'] = base_graph.index + 1
print(base_graph)
```

```python
weighted_graph = pd.read_csv('../results/weighted_income_edge_centrality.csv')
weighted_graph['Weighted Rank'] = weighted_graph.index + 1
print(weighted_graph)
```

```python
merged = pd.merge(base_graph,weighted_graph,on=['Metros'], how='inner')[['Metros','Original Rank', 'Weighted Rank']]
merged['Delta Rank'] = merged['Weighted Rank'] - merged['Original Rank'] 
print(merged)
```

```python
merged.to_csv('../results/link_centrality_changes.csv', index=False)
```
